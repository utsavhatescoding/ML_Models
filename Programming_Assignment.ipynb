```python
def Factorial(num):
  a=1
  for i in range(2,num+1):
    a=a*i
  return a
```

### *iii. Create a program to evaluate permuation and combination*

```python
def Compute_Permutation(n,r):
  Total_Permutations=Factorial(n)/Factorial(n-r)
  return Total_Permutations

def Compute_Combination(n,r):
  Total_Combinations=Factorial(n)/(Factorial(n-r)*Factorial(r))
  return Total_Combinations
```

```python
n=int(input("Enter number of objects : "))
r=int(input("Enter number of arrangments / selection : "))

print("The total permutation is : ",Compute_Permutation(n,r))
print("The total combinations is : ",Compute_Combination(n,r))
```

EMI stands for Equated Monthly Instalment. Essentially, an EMI is a fixed payment amount made by a borrower to a lender on a specified date each month. This instalment includes both the principal loan amount and the interest accrued over the loan tenure.

```python
def EMI_Calculator(P,r,n):
  a=(1+r)**n
  EMI=P*r*(a/(a-1))

  return EMI
```
